# 1) How the Dataset is Obtained

- This is `NASA C-MAPSS turbofan engine degradation` dataset. 
- It is originally obtained from the Nasa's website : https://ti.arc.nasa.gov/tech/dash/groups/pcoe/prognostic-data-repository/
- Kaggle link : https://www.kaggle.com/datasets/behrad3d/nasa-cmaps

# 2) What does the Folder Contains

- This dataset contains run-to-failure simulations of turbofan under different Operational settings (Altitude, Mach number, Throttle resolver angle) and fault modes.
- There are 4 training files (`train_FD001`, `train_FD002`, `train_FD003`, `train_FD004`)
- There are 4 test files (`test_FD001`, `test_FD002`, `test_FD003`, `test_FD004`)
- There are 4 RUL folders (`RUL_FD001`, `RUL_FD002`, `RUL_FD003`, `RUL_FD004`)
- A pdf file containing detailed information about:

    - How the Dataset was Simulated?
    - What are the inputs used to create this data?
    - What are the outputs generated by the model?

### 2.1) Why was the need to create this Dataset on Prognostic Health Management?
- It is difficult to find real world dataset on something as trivial as Prognostic Health Management. Lack of good data, not having the instruments to capture data, and the difficulty to find the dataset that has information from run-to-failure, led the Nasa to build a dataset that will be properly documented, and will contain run-to-failure infomation later to be published publically. 

### 2.2) What is Prognostic Health Management and Why do we need it?

- Here PHM is exclusively considered as *RUL(Remaining Useful Life)* which is curcial for Industries where machineries are used. The real world problem with Machines is, after some particulart amount of time or cycles, it begins to lose its efficiency and in the end it fails. The creates the need of the system which can help you to predict the Remaining Useful life. 
- Changing or Repairing the Machinery while it still has the potential to run, is costly. The alternative, that changing the Machine after it has failed, is not only a waste of efficiency but can lead to halt of production till the machinery is fixed.

### 2.3) How the Dataset was Simulated?
- The Dataset by created by using a Software coded in the MATLAB and Simulink environment. It is a tool for Simulating realistically large turboengine.
- The software includes a number of editable input parameters that allow the user to enter specific values of his/her own choice regarding operational profile, closed-loop controllers, environmental conditions, etc. C-MAPSS simulates an engine model of the 90,000 lb thrust class and the package includes an atmospheric model capable of simulating operations in different conditions

### 2.4) What are the inputs used to create this data?

- The inputs include fuel flow and a set of 13 health-parameter inputs that allow the user to simulate the effects of faults and deterioration in any of the engine’s five rotating components (Fan, LPC, HPC, HPT, and LPT). 

### 2.5) What are the outputs generated by the model?
- The outputs include various sensor response surfaces and operability margins. A total of 21 variables out of  58  different outputs available from the model were used in this study.

# 4) What does Train/Test/RUL file contains?

- **Train** - The Train Dataset consists of run-to-end Data for us to train, with complete cycles from healthy to unoperational.
- **Test** - The Test Dataset contains partial information from healthy state to few steps before engine stopped working.
- **RUL** - The RUL is purely for evaluation purpose, it contains Actual RUL of the test data for us evaluate between Actual and Predicted RUL
